```python
import graphviz as gv
```

```python
def drawDS(ds):
  graph = gv.Digraph("DisjointSet")
  graph.graph_attr['rankdir'] = "BT"
  for e, p in enumerate(ds):
    graph.node(str(e))
    if p >= 0:
      graph.edge(str(e), str(p))
  return graph
```

```python
ds = [-1, -1, -3, 2, 8, -2, 5, -1, -2, 2]
drawDS(ds)
```

# Quick Union

```python
def find(ds, e):
  if ds[e] < 0:
    return e
  else:
    ancestor = find(ds, ds[e])
    ds[e] = ancestor
    return ancestor
```

```python
find(ds, 0), find(ds, 7), find(ds, 3), find(ds, 9), find(ds, 2)
```

```python
def union(ds, a, b):
  a = find(ds,  a)
  b = find(ds, b)

  if a == b: return

  if ds[a] <= ds[b]:
    ds[a] += ds[b]
    ds[b] = a
  else:
    ds[b] += ds[a]
    ds[a] = b
```
